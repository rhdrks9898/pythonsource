### 타입 변환 : astype() vs to_numeric()

```
to_numeric(arg, errors='raise', downcast=None)

arg : scalar, list, tuple, 1-d array, or Serie
errors{‘ignore’, ‘raise’, ‘coerce’}, default ‘raise’
- 'raise' : 변환할 수 없는 값이면 예외 발생
- 'coerce' : 변환할 수 없는 값이면 NaN 으로
- 'ignore' : 변환할 수 없는 값이면 입력 값 그대로

제공되는 데이터 형태에 맞춰 적절하게 숫자 타입으로 변환해 줌 (float64, int64)
```
