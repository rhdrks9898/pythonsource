```python
friend_dict_list = [
    {
        "name":"John",
        "age":25,
        "job":"student"
    },
    {
        "name":"Nate",
        "age":35,
        "job":"teacher"
    },
    {
        "name":"Jenny",
        "age":30,
        "job":"developer"
    },
]

friend_df = pd.DataFrame(friend_dict_list, index=['s1','s2','s3'])
friend_df
```

```python
# and, or : python 
a, b, c = 100,60,15
print("and ", a > b and b > c)
print("or ", a > b or b > c)
```

```python
# 나이가 25 보다 많고, 이름이 Nate 인 사람 조회
# pandas => and : &, or : | 
friend_df[(friend_df['age'] > 25) & (friend_df['name'] == 'Nate')]
```

```python
# query()
# 나이가 25 보다 많고

friend_df.query("age > 25")
```

```python
# 나이가 25 보다 많고, 이름이 Nate 인 사람 조회

friend_df.query("age > 25" and "name == 'Nate'")
```

```python
import seaborn as sns
```

```python
iris_df = sns.load_dataset("iris")
iris_df.head()
```

```python
iris_df.shape
```

> df.iloc[행,열] : 행, 열의 index 값 이용(마지막 값 미포함)

```python
# iloc / loc

# 10 ~ 끝
iris_df.iloc[10:21]

# 인덱스 번호 : 1 ~ 3, 컬럼 : sepal_width, petal_length
iris_df.iloc[1:4,1:3]
```

```python
iris_df.iloc[1:4,[1,2,4]]
```

> df.loc[행,열] : 행, 열의 label 값, boolean 이용(마지막 값 포함)

```python
# 마지막 값 포함

iris_df.loc[1:2]
```

```python
iris_df.loc[1:2,'sepal_width':'petal_width']
```

```python
# 행 1,3  열 : 'sepal_width','petal_width'
# loc
iris_df.loc[[1,3],['sepal_width','petal_width']]
```

```python
iris_df.iloc[[1,3],[1,3]]
```

```python
# sepal_width > 3.0

iris_df[iris_df['sepal_width'] > 3.0]
```

```python
# sepal_width > 3.0, 열 : 'sepal_width','petal_width'

iris_df.loc[iris_df['sepal_width'] > 3.0,['sepal_width','petal_width']]
```

```python
# _ 가 들어있는 컬럼 가져오기
# iris_df.columns

iris_df.filter(regex="_")
```

```python
# length 가 들어있는 컬럼
iris_df.filter(regex="length")

# length 로 끝나는 컬럼
iris_df.filter(regex="length$")

# sepal 로 시작하는 컬럼
iris_df.filter(regex="^sepal")
```
